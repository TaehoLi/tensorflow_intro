## 공통
    - Shift + Enter : 셀 실행 후 아래 셀 선택(없다면 새로 만들기)
    - Ctrl + Enter : 셀 실행
    - Alt + Enter : 셀 실행 후 아래 셀 만들기
    - Esc : 명령 모드
    - Enter : 편집 모드
## 명령모드 (셀 왼쪽 바가 파란색)
    - y / m : 코드 / 마크다운 형식으로 변경
    - Shift + ↑/↓ : 셀 여러개 선택
    - Shift + m : 선택한 셀들을 병합. 하나인 경우 아래와 병합
    - d 2번 누르기 : 셀 삭제
    - c / x / v : 셀 복사 / 잘라내기 / 붙여넣기 (아래에)
    - a / b : 셀을 위에 / 아래에 추가
